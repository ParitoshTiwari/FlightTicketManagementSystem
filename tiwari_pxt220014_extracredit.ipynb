```python
class Flight:
    def __init__(self, id, airline, fromDestination, toDestination, departTime, arrivalTime, totalSeats, cost):
        self.id = id
        self.airline = airline
        self.fromDestination = fromDestination
        self.toDestination = toDestination
        self.departTime = departTime
        self.arrivalTime = arrivalTime
        self.totalSeats = totalSeats
        self.cost = cost
        self.availableSeats = totalSeats
        self.passen = []
    
    def displayFlight(self):
        return f"=================Flight Info==================\nId: {self.id}\nAirline: {self.airline}\nFrom: {self.fromDestination}\nTo: {self.toDestination}\nDeparture Time: {self.departTime}\nArrival Time: {self.arrivalTime}\nAvailable Seats: {self.availableSeats}\nCost: {self.cost}"
```

```python
class Passenger:
    def __init__(self, name, flight = None, amtDue = None):
        self.name = name
        self.flight = flight
        self.amtDue = amtDue
```

```python
class Reservations:
    def __init__(self, passenger, flight):
        self.passenger = passenger
        self.flight = flight
    
    def bookFlight(self):
        self.flight.availableSeats -= 1
        if self.passenger.amtDue is None:
            self.passenger.amtDue = 0
        
        self.passenger.amtDue += self.flight.cost
        self.flight.passen.append(self.passenger)
        print(f"Flight booked for {self.passenger.name}")
        
    def cancelFlight(self):
        self.flight.availableSeats += 1
        self.passenger.amtDue -= self.flight.cost
        self.flight.passen = [item for item in self.flight.passen if item.name != self.passenger.name]
        print("Flight cancelled")
        
    def printTicket(self):
        return f"=================Ticket==================\nPassenger Name: {self.passenger.name}\nAirline: {self.flight.airline}\nFrom: {self.flight.fromDestination}\nTo: {self.flight.toDestination}\nDeparture Time: {self.flight.departTime}\nArrival Time: {self.flight.arrivalTime}\nAvailable Seats: {self.flight.availableSeats}\nCost: {self.flight.cost}"
```

```python
if __name__ == '__main__':
    f1 = Flight(1,'Southwest', 'Dallas', 'Chicago', '6:00 AM', '8:00 AM', 250, 100)
    f2 = Flight(2,'Frontier', 'Boston', 'Chicago', '9:00 AM', '11:00 AM', 300, 110)
    f3 = Flight(3,'Spirit', 'Dallas', 'Chicago', '11:00 AM', '1:00 PM', 90, 125)
    f4 = Flight(4,'Southwest', 'Chicago', 'New York', '2:00 PM', '4:30 PM', 120, 160)
    f5 = Flight(5,'Southwest', 'Dallas', 'California', '4:00 AM', '7:00 AM', 320, 110)
    f6 = Flight(6,'Southwest', 'New York', 'Seattle', '6:00 PM', '9:00 PM', 140, 130)
    f7 = Flight(7,'Southwest', 'Seattle', 'Florida', '1:00 AM', '3:00 AM', 160, 140)
    
    flightList = []
    flightList.append(f1)
    flightList.append(f2)
    flightList.append(f3)
    flightList.append(f4)
    flightList.append(f5)
    flightList.append(f6)
    flightList.append(f7)
    
    while True:
        operation = input("What do you want to do? (Book/Cancel/Reschedule): ")
        name = input("What is your name?: ")
        newPass = Passenger(name)
        if operation.lower() == 'book':
            fromDest = input("From where do you want to book flight?: ")
            toDest = input("To where do you want to book flight?: ")
            availableList = []
            for flight in flightList:
                if flight.fromDestination.lower() == fromDest.lower() and flight.toDestination.lower() == toDest.lower():
                    availableList.append(flight)
                    print(flight.displayFlight())

            checkFlight = False
            bookingFlight = ''
            while True:
                selectedFlightId = int(input("Please select the id of the flight: "))
                items = [item for item in availableList if item.id == selectedFlightId]
                try:
                    bookingFlight = items[0]
                    print(items[0].displayFlight())
                    break
                except:
                    print("No such flight found.")

            confirm = input("Do you want to book this flight? (yes/no): ")
            newPass.flight = bookingFlight
            if confirm.lower() == 'yes':
                reservation = Reservations(newPass, bookingFlight)
                reservation.bookFlight()
        elif operation.lower() == 'cancel':
            checkFound = False
            cancelledFlight = ''
            for flight in flightList:
                for allNames in flight.passen:
                    if allNames.name.lower() == name.lower():
                        cancelledFlight = allNames.flight
                        print("The flight you booked is: ")
                        print(cancelledFlight.displayFlight())
                        checkFound = True
                        break
                if checkFound == True:
                    break
            
            cancel = input('To confirm, enter the id?: ')
            newPass.amtDue = cancelledFlight.cost
            canFlight = Reservations(newPass, cancelledFlight)
            canFlight.cancelFlight()

        elif operation.lower() == 'reschedule':
            checkFound = False
            rescheduleFlight = ''
            for flight in flightList:
                for allNames in flight.passen:
                    if allNames.name.lower() == name.lower():
                        rescheduleFlight = allNames.flight
                        print("The flight you booked is: ")
                        print(rescheduleFlight.displayFlight())
                        checkFound = True
                        break
                if checkFound == True:
                    break
            
            cancel = input('To confirm, enter the id?: ')
            newPass.amtDue = rescheduleFlight.cost
            canFlight = Reservations(newPass, rescheduleFlight)
            canFlight.cancelFlight()
            fromDest = rescheduleFlight.fromDestination
            toDest = rescheduleFlight.toDestination
            availableList = []
            for flight in flightList:
                if flight.fromDestination.lower() == fromDest.lower() and flight.toDestination.lower() == toDest.lower():
                    availableList.append(flight)
                    print(flight.displayFlight())

            checkFlight = False
            bookingFlight = ''
            while True:
                selectedFlightId = int(input("Please select the id of the flight: "))
                items = [item for item in availableList if item.id == selectedFlightId]
                try:
                    bookingFlight = items[0]
                    print(items[0].displayFlight())
                    break
                except:
                    print("No such flight found.")

            confirm = input("Do you want to book this flight? (yes/no): ")
            newPass.flight = bookingFlight
            if confirm.lower() == 'yes':
                reservation = Reservations(newPass, bookingFlight)
                reservation.bookFlight()
            
        nextStep = input("Do you want to continue? (yes/no): ")
        if nextStep.lower() == "no":
            break
```
